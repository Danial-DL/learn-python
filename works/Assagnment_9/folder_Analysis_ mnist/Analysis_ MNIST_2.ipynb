```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#1. Reading the dataset

```python
file_path = '/content/drive/MyDrive/Assagnment_9/folder_Analysis_ mnist/fashion-mnist_test.csv'  # Place the path to your CSV file here.
try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"File {file_path} not found. Please check the file path.")
    exit()
```

# 2. Checking the dimensions of the dataset

```python
num_rows = int(input("Number of rows to display:"))
num_cols = int(input("Number of columns to display:"))

print(f"Number of rows:{df.shape[0]}")
print(f"Number of columns:{df.shape[1]}")
```

# 3. Display a number of rows and columns

```python
print("\n Show part of the dataset:")
print(df.iloc[:num_rows, :num_cols])
```


#4. Dataset Summary

```python
print("\n General data information:")
print(df.info())

print("\n Descriptive statistics of the data set:")
print(df.describe())
```

#5. Drawing charts

## 5.1. Histogram for the 'Labels' column

```python

plt.figure(figsize=(10, 6))
sns.histplot(df['label'], discrete=True)
plt.title('Distribute image stickers')
plt.xlabel('Label')
plt.ylabel('number')
plt.show()
```

## 5.2. Selecting two random columns for a scatter plot

```python
# Given the large size of the dataset, drawing a scatterplot for all columns may be time-consuming.
# Therefore, we choose two random columns.
random_cols = np.random.choice(df.columns[1:], 2, replace=False)  # We are deleting the 'label' column.
plt.figure(figsize=(8, 6))
sns.scatterplot(x=random_cols[0], y=random_cols[1], data=df, alpha=0.5)
plt.title(f'Scatter plot between{random_cols[0]} , {random_cols[1]}')
plt.xlabel(random_cols[0])
plt.ylabel(random_cols[1])
plt.show()
```

## 5.3. Correlation Matrix

```python

# Calculating correlation for the entire dataset can be time-consuming.
# So we use a subset of columns.
num_cols_corr = min(20, df.shape[1])  # Up to 20 columns to calculate correlation
cols_corr = np.random.choice(df.columns[1:], num_cols_corr, replace=False)
cols_corr = np.insert(cols_corr, 0, 'label') # Add a label column
corr_matrix = df[cols_corr].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm")
plt.title('correlation matrix')
plt.show()
```
