```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the data
df = pd.read_csv('/content/drive/MyDrive/Assagnment_9/folder_Analysis_ mnist/fashion-mnist_test.csv')
```

```python
# Show rows and columns
print('Shape of the dataset:', df.shape)
print(df.head())
```

```python
# Number of rows and columns
num_rows, num_columns = df.shape
print(f'There are {num_rows} rows and {num_columns} columns.')
```

```python
# Draw a scatter plot (e.g. based on the first two features)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df.iloc[:, 0], y=df.iloc[:, 1], alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# Draw a histogram
plt.figure(figsize=(15, 10))
df.iloc[:, :-1].hist(bins=30, figsize=(15, 10))
plt.suptitle('Histograms of Features')
plt.show()
```

```python
# Calculate and plot the correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
