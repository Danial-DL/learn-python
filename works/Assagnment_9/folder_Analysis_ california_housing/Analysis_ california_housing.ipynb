##scatter plot

```python
plt.figure(figsize=(10,6))
plt.scatter(data['total_rooms'],data['median_house_value'],alpha=0.5)
plt.title('Scatter Plot of Total Rooms vs Median House Value')
plt.xlabel('Total Rooms')
plt.ylabel('Median House Value')
plt.grid(True)
plt.show()
```

## Histogram

```python
plt.figure(figsize=(10,6))
plt.hist(data['median_house_value'],bins=3,color='blue',alpha=0.7)
plt.title('Histogram of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Freqency')
plt.grid(True)
plt.show()
```

##Correlation

```python
# انتخاب فقط ستون‌های عددی برای محاسبه همبستگی
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# محاسبه همبستگی
correlation_matrix = numeric_data.corr()

# رسم نمودار همبستگی
fig, ax = plt.subplots(figsize=(10, 8))
cax = ax.matshow(correlation_matrix, cmap='coolwarm')

# افزودن برچسب‌ها
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

# افزودن رنگ‌نما
plt.colorbar(cax)

# افزودن عنوان
plt.title("Correlation Matrix", pad=20)

# نمایش نمودار
plt.show()
```
