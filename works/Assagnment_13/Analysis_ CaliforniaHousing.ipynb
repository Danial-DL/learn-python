```python
population__ocean_proximity = data.groupby("ocean_proximity")["population"].sum().reset_index().sort_values(by='population' , ascending=False)
population__ocean_proximity
```

```python
median_house_value__ocean_proximity = data.groupby("ocean_proximity")["median_house_value"].sum().reset_index().sort_values(by='median_house_value' , ascending=False)
median_house_value__ocean_proximity
```

```python
plt.title("Graph of ocean_proximity population ")
sns.lineplot(data=median_house_value__ocean_proximity , x = "ocean_proximity", y="median_house_value")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.title("Graph of ocean_proximity population ")
sns.lineplot(data=population__ocean_proximity , x = "ocean_proximity", y="population")
plt.xticks(rotation=45)
plt.show()
```

```python
Category_population = data.groupby('ocean_proximity')['population'].sum()
Category_population
```

```python
plt.title("ocean_proximity by median_house_value Category")
color = sns.color_palette('pastel')[0:len(Category_population)] # از صفر تا سه رنگ مشخص کرده چون ما سه تا گروه داریم
plt.pie(Category_population,labels=Category_population.index,autopct="%1.1f%%",colors=color)
```

```python
Category_median_house_value = data.groupby('ocean_proximity')['median_house_value'].sum()
Category_median_house_value
```

```python
plt.title("ocean_proximity by median_house_value Category")
color = sns.color_palette('pastel')[0:len(Category_median_house_value)] # از صفر تا سه رنگ مشخص کرده چون ما سه تا گروه داریم
plt.pie(Category_median_house_value,labels=Category_median_house_value.index,autopct="%1.1f%%",colors=color)
```

```python
ocean_proximity_population = data.groupby('ocean_proximity')["population"].sum().reset_index()
ocean_proximity_population
```

```python
plt.title('ocean_proximity by population')
sns.barplot(ocean_proximity_population, x='ocean_proximity' , y='population')
plt.xticks(rotation=75)
```

```python
ocean_proximity_median_house_value =data.groupby('ocean_proximity')["median_house_value"].sum().reset_index()
ocean_proximity_median_house_value
```

```python
plt.title('ocean_proximity by median_house_value')
sns.barplot(ocean_proximity_median_house_value, x='ocean_proximity' , y='median_house_value')
plt.xticks(rotation=75)
```

```python
ocean_proximity = data.groupby("ocean_proximity")[["median_house_value" , "population" , "total_rooms"]].sum().sort_values(by="median_house_value" , ascending=False)
ocean_proximity
```

```python
ocean_proximity.plot(kind="bar")
plt.title("median_house_value and by ocean_proximity")
plt.xlabel("ocean_proximity")
plt.ylabel("median_house_value and population and housing_median_age")
```
