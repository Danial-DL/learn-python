```python
DTYYYYMMDD_HIGH= data.groupby("<DTYYYYMMDD>")["<HIGH>"].sum().reset_index().sort_values(by='<HIGH>' , ascending=False)
DTYYYYMMDD_HIGH
```

```python
plt.title("Graph of DTYYYYMMDD HIGH")
sns.lineplot(data=DTYYYYMMDD_HIGH , x = "<DTYYYYMMDD>" , y = "<HIGH>")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.title("DTYYYYMMDD_HIGH")
sns.barplot(DTYYYYMMDD_HIGH , x='<DTYYYYMMDD>' , y='<HIGH>')
```

```python
plt.title('DTYYYYMMDD_HIGH')
sns.barplot(DTYYYYMMDD_HIGH , x='<DTYYYYMMDD>' , y='<HIGH>')
plt.xticks(rotation=75)
```

```python
TICKER_FIRST = data.groupby('<TICKER>')['<FIRST>'].sum()
```

```python
TICKER_FIRST
```

```python
plt.title("Sales by each TICKER_FIRST")
color = sns.color_palette('pastel')[0:len(TICKER_FIRST)] # از صفر تا سه رنگ مشخص کرده چون ما سه تا گروه داریم
plt.pie(TICKER_FIRST,labels=TICKER_FIRST.index,autopct="%1.1f%%",colors=color)
```
