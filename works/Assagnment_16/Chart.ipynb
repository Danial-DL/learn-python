```python
import matplotlib.pylab as plt 
import numpy as np 
import pandas as pd
```

```python
data = pd.read_csv('housing.csv',encoding='latin-1',)
data.sample(5)
```

```python
housing_median_age = data['housing_median_age']
housing_median_age
```

```python
plt.plot(housing_median_age)
```

```python
total_rooms = data['total_rooms']
total_rooms
```

```python
total_bedrooms = data['total_bedrooms']
total_bedrooms
```

```python
households = data['households']
households
```

```python
median_income = data['median_income']
median_income
```

```python
median_house_value = data['median_house_value']
median_house_value
```

```python
population = data['population']
population
```

```python
longitude = data['longitude']
longitude
```

```python
latitude = data['latitude']
latitude
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.plot(households)
ax2.plot(total_rooms)
ax3.plot(total_bedrooms)
ax4.plot(population)
```

```python
plt.plot(households)
```

```python
plt.plot(households.cumsum())
```

```python
plt.plot(households.cumsum(),'k--')
```

```python
plt.plot(median_house_value)
```

```python
plt.plot(median_house_value.cumsum())
```

```python
plt.plot(median_house_value.cumsum(),'k--')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.hist(median_house_value,bins=30 , color='k' , alpha=0.5)
ax2.scatter(longitude,latitude)
ax3.plot(median_income)
ax4.plot(housing_median_age)
```

```python
fig , axes =plt.subplots(2,2, sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500), bins=50 , color = 'r' , alpha=0.6)
plt.subplots_adjust(wspace=0 , hspace=0)
```

```python
from numpy.random import randn
plt.plot(randn(30).cumsum() , 'ko--')
```

```python
plt.plot(randn(30).cumsum() , color= 'g' , linestyle='dashed' , marker='o')
```

```python
data = randn(30).cumsum()
plt.plot(data , 'k--' , label='Default')
plt.plot(data , 'ko-' , drawstyle='steps-post', label='steps-post')

plt.legend(loc='best')
```
